# Analyzing top countries by more customers and sales 

```python
query1 = '''  SELECT C.COUNTRY,    COUNT( DISTINCT C.CUSTOMERID) as numÙ€customers  ,sum(total)
  FROM CUSTOMER C
  JOIN INVOICE I ON C.CUSTOMERID = I.CUSTOMERID
  GROUP BY COUNTRY 
  ORDER BY numÙ€customers DESC;'''
top_countries = pd.read_sql_query(query1 ,connection)
```

```python
top_countries.head()
```

```python


plt.figure(figsize=[20,8])

plt.suptitle('Top Countries',fontsize = 24)

plt.subplot(1,2,1) 
plt.bar(top_countries['Country'][0:5],top_countries['sum(total)'][0:5])
plt.title(' Number of Total Sales By Country',fontsize = 20)
plt.ylabel('Number of Total Sales',fontsize = 16)
plt.xticks( fontsize = '14', family = 'fantasy',color='k');
plt.yticks( fontsize = '18', family = 'fantasy',color='k');


plt.subplot(1,2,2)
plt.bar(top_countries['Country'][0:5],top_countries['numÙ€customers'][0:5])
plt.ylabel('Number of Customers',fontsize = 16)
plt.title('Number of Customers By Country',fontsize = 20);
plt.xticks( fontsize = '14', family = 'fantasy',color='k');
plt.yticks( fontsize = '18', family = 'fantasy',color='k');

```

```python


sns.relplot(x=top_countries['numÙ€customers'][0:5],
y=top_countries['sum(total)'][0:5],
hue=top_countries['Country'][0:5],
size=top_countries['numÙ€customers'][0:5], sizes=(50,500))
plt.xlabel('Number of Customers',fontsize = 16)
plt.ylabel('Number of Total Sale',fontsize = 16)




```

# Identifying Popular Music Genres



```python
query2 = '''  SELECT DISTINCT  G.NAME  ,COUNT(  G.NAME)
  FROM CUSTOMER C
  JOIN INVOICE I ON C.CUSTOMERID = I.CUSTOMERID
  JOIN INVOICELINE IL ON IL.INVOICELINEID = I.INVOICEID
  JOIN TRACK T ON IL.TRACKID = T.TRACKID
  JOIN GENRE G ON T.GENREID = G.GENREID
   GROUP BY G.NAME
  ORDER BY COUNT(  G.NAME) DESC;'''
top_music= pd.read_sql_query(query2 ,connection)
```

```python
top_music.describe()
```

```python
top_music.head()
```

```python
plt.figure(figsize=[20,8])
plt.suptitle('Top Popular Music',fontsize = 24)

plt.subplot(1,2,1) 
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#66b3ff']
plt.pie(top_music['COUNT(  G.NAME)'][0:5],labels= top_music['Name'][0:5],colors=colors);
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.figure(figsize=[20,8])


plt.subplot(1,2,2) 
plt.bar(top_music['Name'][0:5],top_music['COUNT(  G.NAME)'][0:5], color = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#66b3ff'])
plt.title(' Top Popular Music By sales',fontsize = 20)
plt.ylabel('Sales',fontsize = 16)
plt.xticks( fontsize = '14', family = 'fantasy',color='k');
plt.yticks( fontsize = '18', family = 'fantasy',color='k');


```

# Identifying Popular Artist by Top Music Genres



```python
query3 = '''SELECT ar.name artistname, COUNT(al.albumid) albums_sold
FROM  Artist  ar
INNER JOIN album al ON ar.artistid = al.artistid
INNER JOIN track t ON t.albumid = al.albumid
INNER JOIN genre g ON t.genreid = g.genreid
WHERE g.name = "Rock" or "Latin" or "Metal" or "Alternative & Punk" or "Jazz"
GROUP BY  artistname
ORDER BY albums_sold DESC
LIMIT 5;'''
top_artist= pd.read_sql_query(query3 ,connection)
```

```python
top_artist.head()
```

```python
plt.grid()
plt.plot(top_artist['artistname'],top_artist['albums_sold'])
plt.scatter(top_artist['artistname'],top_artist['albums_sold']);
plt.xticks( fontsize = '10', family = 'fantasy',color='gray');
plt.yticks( fontsize = '10', family = 'fantasy',color='gray');
```
